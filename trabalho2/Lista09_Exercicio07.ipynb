##Exercicio 07 - Lista 09 (K-Means)

###Importação de bibliotecas

```python
# Importando as bibliotecas necessárias.
from imageio import imread
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
```

###Abrindo a Imagem

```python
imagem = imread("img2.jpg")

#Exibindo as dimensões da array carregada
print(imagem.shape)
```

###Exibindo a Imagem

```python
plt.figure()
plt.imshow(imagem)
```

###Conversão em array 2D

```python
#Load Image and transform to a 2D numpy array.
w, h, d = original_shape = tuple(imagem.shape)
assert d == 3
imagem = np.array(imagem, dtype=np.float64) / 255
image_array = np.reshape(imagem, (w * h, d))

print(image_array.shape)
```

###Treinamento K-Means para 2 Clusters

```python
amostras2 = shuffle(image_array)[:1000]

kmeans_2 = KMeans(n_clusters=2, random_state = 42)

kmeans_2.fit(amostras2)

predict_2 = kmeans_2.predict(image_array)

for i in range(0,2):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,kmeans_2.cluster_centers_[i][0],kmeans_2.cluster_centers_[i][1],kmeans_2.cluster_centers_[i][2]))
```

###Treinamento K-Means para 4 Clusters

```python
amostras4 = shuffle(image_array)[:1000]

kmeans_4 = KMeans(n_clusters=4, random_state = 42)

kmeans_4.fit(amostras4)

predict_4 = kmeans_4.predict(image_array)

for i in range(0,4):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,kmeans_4.cluster_centers_[i][0],kmeans_4.cluster_centers_[i][1],kmeans_4.cluster_centers_[i][2]))
```

###Treinamento K-Means para 8 Clusters

```python
amostras8 = shuffle(image_array)[:1000]

kmeans_8 = KMeans(n_clusters=8, random_state = 42)

kmeans_8.fit(amostras8)

predict_8 = kmeans_8.predict(image_array)

for i in range(0,8):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,kmeans_8.cluster_centers_[i][0],kmeans_8.cluster_centers_[i][1],kmeans_8.cluster_centers_[i][2]))
```

###Treinamento K-Means para 16 Clusters

```python
amostras16 = shuffle(image_array)[:1000]

kmeans_16 = KMeans(n_clusters=16, random_state = 42)

kmeans_16.fit(amostras16)

predict_16 = kmeans_16.predict(image_array)

for i in range(0,16):
    print('Centroid %d: (%1.2f, %1.2f, %1.2f)' % (i,kmeans_16.cluster_centers_[i][0],kmeans_16.cluster_centers_[i][1],kmeans_16.cluster_centers_[i][2]))
```

###Função recriando a imagem

```python
def recreate_image(codebook, labels, w, h):
    """Recreate the (compressed) image from the code book & labels"""
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    #For each pixel setting the color of its nearest cluster mean
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image
```

###Recriando as 4 iamgens

```python
## 2 Clusters
plt.figure(1)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(kmeans_2.cluster_centers_, predict_2, w, h))
plt.title('2 CORES')

## 4 Clusters
plt.figure(2)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(kmeans_4.cluster_centers_, predict_4, w, h))
plt.title('4 CORES')

## 8 Clusters
plt.figure(3)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(kmeans_8.cluster_centers_, predict_8, w, h))
plt.title('8 CORES')

## 16 Clusters
plt.figure(4)
plt.clf()
ax = plt.axes([0, 0, 1, 1])
plt.axis('off')
plt.imshow(recreate_image(kmeans_16.cluster_centers_, predict_16, w, h))
plt.title('16 CORES')
```

###Tamanho aproximado das imagens



```python
pixels = 400 * 500
#Considerando que há um total de 24 bits para representar cada cor de cada pixel,
#será feito os calculos de 1bit para 2 cores, 2 bits para 4 cores, 3 bits para 8 cores e 4 bits para 16 cores.

original = pixels * 3 #bytes por pixel

print ("Imagem original: ", original/1000, " kBytes")

#A divisão por 8 abaixo está sendo feita para converter o resultado para bytes
cores2 = pixels * 1 / 8
cores4 = pixels * 2 / 8
cores8 = pixels * 3 / 8
cores16 = pixels * 4 / 8

print ("Imagem 2 cores: ", cores2/1000, " kBytes")
print ("Imagem 4 cores: ", cores4/1000, " kBytes")
print ("Imagem 8 cores: ", cores8/1000, " kBytes")
print ("Imagem 16 cores: ", cores16/1000, " kBytes")
```

### É possível concluir que com a diminuição de cores, as imagens podem permanecer semelhantes, diminuindo seu tamanho. Durante o desenvolvimento do exercício também foi verificado com 32 e 64 clustes, onde foi possível ver a semelhança entre elas.
