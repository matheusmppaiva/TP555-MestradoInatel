```python
import pandas as pd
import json
import os
```

```python
num_rows = 3200000
```

```python
base_folder = 'D:/Workspace/IoT-23 Dataset/iot_23_datasets_small/opt/Malware-Project/BigDataset/IoTScenarios/'
sub_folders = filter(lambda x : ('Honeypot' not in x and 'bro' in x), [x[0] for x in os.walk(base_folder)])
base_filename = 'conn.log.labeled'
```

```python
data_frames = []

for folder in sub_folders:  
    
    full_filename = folder.replace('\\','/') + '/' + base_filename
    
    df = pd.read_table(filepath_or_buffer=full_filename, skiprows=8, nrows=num_rows)
    
    df.columns = [
        'ts',
        'uid',
        'id.orig_h',
        'id.orig_p',
        'id.resp_h',
        'id.resp_p',
        'proto',
        'service',
        'duration',
        'orig_bytes',
        'resp_bytes',
        'conn_state',
        'local_orig',
        'local_resp',
        'missed_bytes',
        'history',
        'orig_pkts',
        'orig_ip_bytes',
        'resp_pkts',
        'resp_ip_bytes',
        'label'
    ]
    
    df.drop(columns=['ts','uid','service','local_orig','local_resp','history'], inplace=True)
    
    df.drop(df.tail(1).index, inplace=True)
    
    data_frames.append(df)
    
    print(f'Folder \'{folder}\' proccessed successfully; DF shape: {df.shape}.')
```

```python
df_c = pd.concat(data_frames)
```

```python
def filter_value_counts(df,col,pct=0.99):
    print('----------------------------------------')
    vc = df[col].value_counts()
    print(f'**** Value counts (before) ****\n{vc}')
    threshold = vc.quantile(pct)
    print(f'\n**** Dropping \'{col}\' rows with less than {threshold:.2f} occurrences ****\n')
    df = df[df[col].isin(vc.index[vc.gt(threshold)])]
    print(f'**** Value counts (after) ****\n{df[col].value_counts()}')
    return df
```

```python
to_be_filtered = ['id.orig_h','id.orig_p','id.resp_h','id.resp_p']
for col in to_be_filtered:
    df_c = filter_value_counts(df_c,col)
    print(f'**** Column \'{col}\' successfully truncated; new DF shape: {df_c.shape} ****\n')
```
