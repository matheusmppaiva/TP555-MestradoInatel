```python
def filter_value_counts(df,col,pct=0.99):
    vc = df[col].value_counts()
    print(f'**** Value counts (before) ****\n{vc}')
    threshold = vc.quantile(pct)
    print(f'\n**** Dropping \'{col}\' rows with less than {threshold:.2f} occurrences ****\n')
    df = df[df[col].isin(vc.index[vc.gt(threshold)])]
    print(f'**** Value counts (after) ****\n{df[col].value_counts()}')
    print('----------------------------------------\n')
    return df
```

```python
for col in ['id.orig_h','id.orig_p','id.resp_h','id.resp_p']:
    df_c = filter_value_counts(df_c,col)
```
